```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(df.iloc[:,:-1],df['Outcome'],test_size=0.3,random_state=16)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state=16)
```

```python
logreg.fit(X_train,y_train)
```

```python
y_train.head(2)
```

```python
y_pred = logreg.predict(X_test)
ans=pd.DataFrame({'y_pred':y_pred,'y_actual':y_test})
```

```python
len(ans[ans['y_pred']!=ans['y_actual']]),len(ans[ans['y_pred']==ans['y_actual']]),len(ans)
```

```python
from sklearn import metrics
```

```python
cnf_mat = metrics.confusion_matrix(y_test,y_pred)
```

```python
cnf_mat
```

```python
target_names = ['without diabetes', 'with diabetes']
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred,target_names=target_names))
```
