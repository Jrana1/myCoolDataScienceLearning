# Data Normalization
**Normalization** is the process of transforming values of several variables into a similar range. Typical normalizations include scaling the variable so that variable **avg** is 0, scalling the variable so the **variance** is 1. or scalling the variable so the variable **values** ranges from 0 to 1. As an example consider the following data set.


| age | income |
|----------|----------|
| 20  | 100000   | 
| 30   | 20000  | 
| 40   | 250000  | 


**Age** and **income** are in different range. They are hard to compare. **income** will influence the result(for example in linear regression) more. After **normalizaton** the data set is..

| age | income |
|----------|----------|
| .2  | .2   | 
| .3   | .04  | 
| .4   | 1  | 

## There are severals methods to normalize data

+ Simple Feature scalling
   $$X_{new}=\frac{X_{old}}{X_{max}}$$
+ Min-Max
    $$X_{new}=\frac{X_{old} - X_{min}}{X_{max}-X_{min}}$$
+ Z-Score
    $$X_{new}=\frac{X_{old} -\mu}{\sigma}$$
    z values ranges from $+3$ to $-3$

Normalize **height** using **simple z-score methods**

```python
df['height'].plot()
```

```python
df['height'] = (df['height'] - df['height'].mean())/df['height'].var()
df['height']
```

# Binning
Binning is a process of transforming continuous numerical variables into descrete categorical 'bins' to grouped analysis.<br/>
**Example:**<br/>
<p>In our dataset, "horsepower" is a real valued variable ranging from 48 to 288 and it has 59 unique values. What if we only care about the price difference between cars with high horsepower, medium horsepower, and little horsepower (3 types)? Can we rearrange them into three ‘bins' to simplify analysis? </p>

<p>We will use the pandas method 'cut' to segment the 'horsepower' column into 3 bins.</p>
<h3>Example of Binning Data In Pandas</h3>


```python
df['horsepower']=df['horsepower'].astype('float')
df['horsepower'].info()
```

```python
df['horsepower'].replace(np.nan,df['horsepower'].mean(),inplace=True)
df['horsepower'].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df['horsepower'])
plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

<p>We would like 3 bins of equal size bandwidth so we use numpy's <code>linspace(start_value, end_value, numbers_generated)</code> function.</p>
<p>Since we want to include the minimum value of horsepower, we want to set start_value = min(df["horsepower"]).</p>
<p>Since we want to include the maximum value of horsepower, we want to set end_value = max(df["horsepower"]).</p>
<p>Since we are building 3 bins of equal length, there should be 4 dividers, so numbers_generated = 4.</p>


```python
bins = np.linspace(min(df['horsepower']),max(df['horsepower']),4)
bins
```

```python
group_names=['low','medium','high']
```

```python
df['horsepower-grouped']=pd.cut(df['horsepower'],bins,labels=group_names,include_lowest=True)
df[['horsepower','horsepower-grouped']].head()
```

```python
df['horsepower'].value_counts()
```

```python
%matplotlib inline
pyplot.bar(group_names,df['horsepower-grouped'].value_counts())
```

# Indicator Variable(or Dummy Variable)
## what is an indicator variable?
An indicator variable( or dummy) is a numerical variable used to label categories. They are called 'dummies' because the numbers themselves don't have inherent meaning.
## why to use them?
We use dummy variable sothat we can use categorical variables for regression analysis in the later modules.
<br/>
<b>Example</b>
<p>
    We see the column "fuel-type" has two unique values: "gas" or "diesel". Regression doesn't understand words, only numbers. To use this attribute in regression analysis, we convert "fuel-type" to indicator variables.
</p>

<p>
    We will use pandas' method 'get_dummies' to assign numerical values to different categories of fuel type. 
</p>


```python
dummy_variable=pd.get_dummies(df['fuel-type'])
dummy_variable
```

```python
dummy_variable.rename(columns={'gas':'fuel-type-gas','diesel':'fuel-type-diesel'},inplace=True)
dummy_variable
```

```python
df=pd.concat([df,dummy_variable],axis=1)
df.head()
```

```python
df.drop('fuel-type',axis=1,inplace=True)
df.head()
```
